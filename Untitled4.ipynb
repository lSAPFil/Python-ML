```python
df_train=df_train.drop(columns="YEAR", axis=1)
df_train=df_train.drop(columns="MONTH", axis=1)
df_train=df_train.drop(columns="DAY_OF_WEEK", axis=1)
df_train=df_train.drop(columns="Lat", axis=1)
df_train=df_train.drop(columns="Long", axis=1)
df_train=df_train.drop(columns="Location", axis=1)
df_train=df_train.drop(columns="HOUR", axis=1)
df_train=df_train.drop(columns="OFFENSE_CODE_GROUP", axis=1)
df_train=df_train.drop(columns="OFFENSE_DESCRIPTION", axis=1)
df_train=df_train.drop(columns="INCIDENT_NUMBER", axis=1)
df_train=df_train.drop(columns="DISTRICT", axis=1)
```
