# Предобработка данных и выделение значимых атрибутов

```python
df = pn.DataFrame({"Название_компетенции":[],
                   "Описание_компетенции":[],
                   "Название_раздела":[],
                   "Значимость_раздела":[],
                   "Признаки_раздела":[],})
for i in range(0,27):
    df["column_"+str(i)]=0
    
for i in df.columns:
    df[i]=df[i].astype(str)
    
data.columns=df.columns
```

```python
data.info()
```

```python
# В базе имеется большое кол-во смешанных данных, их нужно разделить по столбцам
# чтобы было красивенько
# Мы перезаписываем всю таблицу в новую таблицу с более упорядоченными значениями (крч наводим красоту)
# Мы до того создали столбцы в которых как раз все добавляем в самом конце
# В столбцы уже должны быть упорядоченные данные. 
# Во всей базе данные можно разделить на 5 классификаций которые упорядоченно разбросаны по 28 столбцах
# создаем таблицы q, w, e и вписываем их построчно в нужный столбец снизу


index_row=-1
# Работа range(От, До, Шаг) => работа range идет как range(от 2, до 28, шагом в 3) и записывается в первую переменную q

for i in data.itertuples():
    index_row=index_row+1
    for q,w,e in zip(range(2,28,3), range(3,28,3),range(4,28,3)):
        #Шаг в 3 потому что сама БД строится через 3 
        #(умения спеца, язык, баллы)
        if pn.isnull(data.iloc[index_row][q]):
             #Остановка, потому что если нет q то и нет w и e
            break
        else:
            df.loc[len(df)]=np.nan
            # останавливается на столбце, если есть NaN
            df = df.shift()
            # увидел NaN и сдвигается на шаг(компандой Shift), после чего
            # вставляет нижние столбцы
            df.at[0,"Название_компетенции"]=data.iloc[index_row][0]
            df.at[0,"Описание_компетенции"]=data.iloc[index_row][1]
            df.at[0,"Название_раздела"]=data.iloc[index_row][q]
            df.at[0,"Значимость_раздела"]=data.iloc[index_row][w]
            df.at[0,"Признаки_раздела"]=data.iloc[index_row][e]
            
df.drop(df.iloc[:,5:], inplace=True, axis=1)
```

## Очистка данных

- Перевести слова в нижний регистр
- Удалить все, что не является русским или английскими словами, пунктуационные знаки и т.д
- Удалить стоп-слова
- Удалить лишнеее пространства между словами

```python
import nltk
from nltk.corpus import stopwords

nltk.download("stopwords")
```

```python
def clean_data(df):
    df = df.str.lower()
    df = df.replace(r'[^а-яА-Яa-zA-Z]', ' ', regex = True)
    df = df.str.strip()
    stop_words = stopwords.words("russian")
    df = df.apply(lambda x: ' '.join([word for word in x.split() 
                                      if word not in (stop_words)]))
    return df
```

```python
df["Признаки_раздела"]=clean_data(df["Признаки_раздела"])
```

# Feature Engineering

```python
df["count_description_len"] = np.nan
df["count_title_len"] = np.nan
df["count_titlefeat_len"] = np.nan

df = df.reset_index()

for row in range(len(df)):
    df["count_description_len"][row] = len(df["Описание_компетенции"][row])
    
for row in range(len(df)):
    df["count_title_len"][row] = len(df["Название_раздела"][row])

for row in range(len(df)):
    df["count_titlefeat_len"][row] = len(df["Признаки_раздела"][row])
    
df["count_description_len"] = df["count_description_len"].astype(int)
df["count_title_len"] = df["count_title_len"].astype(int)
df["count_titlefeat_len"] = df["count_titlefeat_len"].astype(int)
```

```python
# Трансформаторы для получения данных
from sklearn.base import BaseEstimator, TransformerMixin
# Конвеер для объединения шагов выполнения классификации
from sklearn.pipeline import Pipeline
# Функция стандартизации числовых значений
from sklearn.preprocessing import StandardScaler
# Функция преобразования текстовых данных под модель
from sklearn.feature_extraction.text import TfidfVectorizer

# Класс необходимый для получения и работы с текстовыми признаками
class TextSelector(BaseEstimator, TransformerMixin):
    def __init__(self, key):
        self.key = key
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.key]
    
# Класс необходимый для получения работы с числовыми признаками
class NumericalSelector(BaseEstimator, TransformerMixin):
    def __init__(self, key):
        self.key = key
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[[self.key]]
```

```python
dictionary = Pipeline([
    ('selector', TextSelector(key="Описание_компетенции")),
    ('tfidf', TfidfVectorizer(stop_words = stop_words_b))
    ])

dictionary.fit_transform(df).todense()
```
